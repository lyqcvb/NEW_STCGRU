```python
import mne
import matplotlib.pyplot as plt

# 定义脑区和对应的电极（使用实际通道名）
regions = {
    "Frontal": ["EEG Fp1-Ref", "EEG Fp2-Ref", "EEG F3-Ref", "EEG F4-Ref", "EEG F7-Ref", "EEG F8-Ref", "EEG Fz-Ref"],
    "Central": ["EEG C3-Ref", "EEG C4-Ref", "EEG Cz-Ref"],
    "Temporal": ["EEG T3-Ref", "EEG T4-Ref", "EEG T5-Ref", "EEG T6-Ref"],
    "Parietal": ["EEG P3-Ref", "EEG P4-Ref", "EEG Pz-Ref"],
    "Occipital": ["EEG O1-Ref", "EEG O2-Ref"]
}

# 加载 EEG 数据
raw = mne.io.read_raw_edf('糖尿病认知障碍与对照脑电数据\认知障碍\陈艳杰.edf',encoding="latin1",preload=True,verbose='Warning')

# 按脑区分类数据
region_data = {}
for region, electrodes in regions.items():
    selected_channels = [ch for ch in electrodes if ch in raw.ch_names]
    if selected_channels:
        region_data[region] = raw.copy().pick_channels(selected_channels)

# 可视化每个脑区的功率谱密度
for region, data in region_data.items():
    psd = data.compute_psd(fmin=0.5, fmax=32, n_fft=1024)  # 替代 psd_welch
    psd_data, freqs = psd.get_data(return_freqs=True)    # 获取 PSD 数据和频率
    plt.plot(freqs, psd_data.mean(axis=0), label=f"{region} PSD")

plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density (uV^2/Hz)")
plt.legend()
plt.title("Power Spectral Density by Region")
plt.show()

```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.signal import welch

# 假设 EEG 信号是一个形状为 (19, 20000) 的 numpy 数组
# 19 代表通道数，20000 代表每个通道的时间点数
# 生成随机的EEG信号数据作为示例
eeg_data = np.random.randn(19, 20000)

# 时域特征计算
def calculate_time_domain_features(data):
    features = {}
    features['mean'] = np.mean(data, axis=1)
    features['std'] = np.std(data, axis=1)
    features['skew'] = stats.skew(data, axis=1)
    features['kurtosis'] = stats.kurtosis(data, axis=1)
    return features

# 频域特征计算
def calculate_frequency_domain_features(data, fs=500):
    features = {}
    n_channels = data.shape[0]
    f, Pxx = welch(data[0, :], fs=fs, nperseg=1024)  # 计算第一个通道的功率谱密度
    features['psd'] = Pxx
    
    # 计算频带功率（例如 Delta, Theta, Alpha, Beta）
    def band_power(f, Pxx, band):
        idx_band = np.logical_and(f >= band[0], f <= band[1])
        return np.trapz(Pxx[idx_band], f[idx_band])
    
    bands = {'delta': (0.5, 4), 'theta': (4, 8), 'alpha': (8, 13), 'beta': (13, 30)}
    for band_name, band_range in bands.items():
        features[band_name] = band_power(f, Pxx, band_range)

    # 频谱熵计算
    def spectral_entropy(Pxx):
        Pxx_norm = Pxx / np.sum(Pxx)
        return -np.sum(Pxx_norm * np.log2(Pxx_norm))
    
    features['spectral_entropy'] = spectral_entropy(Pxx)
    
    return features

# 计算时域特征
time_domain_features = calculate_time_domain_features(eeg_data)

# 计算频域特征
frequency_domain_features = calculate_frequency_domain_features(eeg_data)

# 输出结果
print("Time Domain Features (Mean, Std, Skewness, Kurtosis):")
print(time_domain_features)

print("\nFrequency Domain Features (PSD, Band Power, Spectral Entropy):")
print(frequency_domain_features)

# 可视化某个通道的功率谱密度
plt.figure(figsize=(10, 6))
plt.semilogy(frequency_domain_features['psd'])
plt.title("Power Spectral Density of Channel 1")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power/Frequency (dB/Hz)")
plt.grid(True)
plt.show()
```
